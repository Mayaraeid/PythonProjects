```python
# 1. Algoritmos de Árvores de Decisão e Ensemble de Árvores
tree_and_ensemble = {
    'RF': 15,
    'DT': 5,
    'XGBoost': 2,
    'AdaBoost': 1,
    'Boosted and Bagged Decision Tree': 1,
    'Boosted and Bagged Regression Tree': 1,
    'Extra Trees Bagging': 1,
    'eXtreme Gradient Boosting': 1,
    'Gradient Boosting': 1
}

# 2. Redes Neurais e Arquiteturas Convolucionais
nn_e_cnn = {
    'MLP': 10,
    'NN': 4,
    'CNN': 2,
    'ResNet50': 2,
    'VGG-16': 2,
    'Xception': 2,
    'AIPB + QCNN': 1,
    'Alex-Net': 1,
    'DNN': 1,
    'Inception': 1,
    'Inception V3': 1,
    'ResNet50V2': 1,
    'ResNetV2': 1,
    'SkipConNet': 1,
    'SqueezeNet': 1
}

# 3. Modelos Baseados em Métodos Estatísticos e Probabilísticos
statistical_and_probabilistic_methods = {
    'LR': 6,
    'NB': 6,
    'fGMM': 1,
    'Gaussian Process': 1,
    'GMM-fUBM': 1,
    'Linear Discriminant': 1,
    'opt-fGMM-fUBM': 1,
    'Quadratic Discriminant': 1
}

# 4. Redes Neurais Recorrentes e LSTM
recurrent_networks_and_lstm = {
    'LSTM': 1,
    'NF': 1
}

# 5. Algoritmos Baseados em Support Vector Machines
svm_group = {
    'SVM': 20,
    'SVR': 2
}

# 6. Algoritmos Baseados em Métodos Estocásticos e Outros
stochastic_and_neighborhood_methods_and_others = {
    'KNN': 9,
    'Stochastic Gradient': 1,
    'Two-Stage Variable': 1
}

# 7. Ensemble e Aprendizado Combinado
ensemble_and_combined_learning = {
    'DA': 1,
    'Ensemble Learning': 1,
    'Forest Net': 1
}

```

```python
import pandas as pd
import matplotlib.pyplot as plt

algorithm_group_db = {
    'Trees and Ensemble': tree_and_ensemble,
    'Neural Networks and CNN': nn_e_cnn,
    'Statistical and Probabilistic Methods': statistical_and_probabilistic_methods,
    'Recurrent Networks and LSTM': recurrent_networks_and_lstm,
    'SVM Group': svm_group,
    'Stochastic and Neighborhood Methods and Others': stochastic_and_neighborhood_methods_and_others,
    'Ensemble and Combined Learning': ensemble_and_combined_learning
}

# Criar um dataframe
df_list = []
for category, algorithms in algorithm_group_db.items():
    for algorithm, count in algorithms.items():
        df_list.append([category, algorithm, count])

df = pd.DataFrame(df_list, columns=['Category', 'Algorithm', 'Count'])

total_count = df['Count'].sum()

df['Category_Percentage'] = df.groupby('Category')['Count'].transform('sum') / total_count * 100

df['Algorithm_Percentage_Within_Category'] = df.groupby('Category')['Count'].transform(lambda x: x / x.sum() * 100)

df['Algorithm_Percentage_Total'] = df['Count'] / total_count * 100

df = df.sort_values(by=['Category_Percentage', 'Algorithm_Percentage_Within_Category'], ascending=[False, False])

plt.figure(figsize=(14, 10))
pivot_df = df.pivot(index='Category', columns='Algorithm', values='Algorithm_Percentage_Total').fillna(0)


# Função para gerar tons de azul
def generate_blue_palette(num_colors):
    return plt.cm.BuGn(np.linspace(0.8, 0.1, num_colors))

blue_palette = generate_blue_palette(42)

blue_palette_hex = [mcolors.to_hex(color) for color in blue_palette]

#ax = pivot_df.plot(kind='bar', stacked=True, figsize=(14, 10), color=blue_palette_hex[:len(pivot_df.columns)], width=0.7)
ax = pivot_df.plot(kind='bar', stacked=True, figsize=(14, 10), color=blue_palette_hex[:len(pivot_df.columns)], width=0.7, edgecolor='black', linewidth=0.05)

for i, category in enumerate(pivot_df.index):
    bottom = 0
    for algorithm in pivot_df.columns:
        value = pivot_df.loc[category, algorithm]
        if pd.notna(value) and value > 0:  # Garantir que o valor é finito e positivo
            label = f'{algorithm}' if value < 0.5 else f'{algorithm}\n{value:.1f}%'
            ax.text(i, bottom + value / 2, label, ha='center', va='center', fontsize=6, color='black')
        bottom += value

category_totals = df.groupby('Category')['Category_Percentage'].max()
for i, (category, total) in enumerate(category_totals.items()):
    feature_count = df[df['Category'] == category].shape[0]  # Contar o número de features em cada categoria
    if pd.notna(total):  # Garantir que o total é finito
        ax.text(i, total + 1, f'{total:.1f}%\nFeatures: {feature_count}', ha='center', va='bottom', fontsize=8, color='black')

plt.title('Percentage Distribution of Algorithms within Each Category', fontsize=12, fontweight='bold')
plt.ylabel('Percentage (%)', fontsize=10)
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.ylim(0, 30)  # Limitar o eixo y a 25%
plt.grid(axis='y', linestyle='--', alpha=0.7)
ax.legend_.remove()  # Remover a legenda externa

plt.tight_layout()
plt.savefig('algorithm_group.png', dpi=300, bbox_inches='tight')  # PNG com alta resolução
plt.show()
```
